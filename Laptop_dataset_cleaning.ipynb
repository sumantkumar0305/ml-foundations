## OpOS column



1.   Make a seprate columns for every type of OS like 'Windows 10' and 'Windows 7' -> 'WindowsOS', 'macOS' and 'Mac OS X' -> 'MacOS', 'Linux' -> 'LinuxOS', 'Chrome OS' and 'Andorid' -> 'Other' and 'No OS' -> 'No_OS'

2. All this new columns only contain 0 or 1. 0 mean device not belong to the OS and 1 mean device belong to the OS

3. Finally delete this columns




```python
df['OpSys'].unique()
```

```python
df.head()
```

```python
df['OpSys'].unique()
```

```python
df['OpSys'] = df['OpSys'].str.replace('Windows 10 S', 'Windows 10')
```

```python
df['Windows'] = df['OpSys'].str.startswith('Windows').astype('int')
```

```python
df['Mac'] = df['OpSys'].str.startswith('macOS').astype('int')
```

```python
df['Linux'] = (df['OpSys'] == 'Linux').astype('int')
```

```python
df['OtherOS'] = ((df['OpSys'] == 'Chrome OS') | (df['OpSys'] == 'Android')).astype('int')
```

```python
df.rename(columns={'Windows': 'WindowsOS'}, inplace=True)
```

```python
df.rename(columns={'Mac': 'MacOS'}, inplace=True)
```

```python
df.rename(columns={'Linux': 'LinuxOS'}, inplace=True)
```

```python
df['No_OS'] = (df['OpSys'] == 'No OS').astype('int')
```
